###54.
Crear una función que grafique la derivada de otra función, pasándole como argumentos la función y el rango de valores para el cual quiere calcular la derivada. Utilice como base el código de la celda Calc01

```python
def derivative(fx,n=1,n1=0,n2=10):
  """ 
  Función que calcula la derivada grado n de una función f(x) dada y la plotea con unos valores de x dados. 
  Argumentos: 
  fx -- función en expresión considerando el simbolo "x" como parametro que se va a derivar.
  n -- grado de la derivada. 
  n1 -- valor minimo de ploteo variable x.
  n2 -- valor maximo de ploteo variable x.
  """
  import numpy as np
  import scipy as sp
  import sympy as syp
  import matplotlib.pyplot as plt
  x= syp.symbols('x')
  expr=syp.Derivative(fx,x,n).doit()
  f = syp.lambdify(x, expr, "numpy") 
  r = np.arange(n1,n2+1)
  f_x=np.asarray(f(r))
  if (f_x.size)==1:
    f_x=np.full(r.shape,f_x,)
  plt.plot(r,f_x,'-.r'); ax=plt.gca(); ax.set_xlim([n1,n2])
  return f_x
  
import sympy as syp

syp.var('x')
fx=10+5*x+x**3
dnfx=derivative(fx,1)

help(derivative)
```
